```python
import pandas as pd
```

```python
stats = pd.read_csv('../data/Pacific.csv')
stats
```

```python
# Most aggresive team (first kill/ (first death + first kill))
result = stats.groupby(by='Team')[['Rnd', 'FK', 'FD',]].sum()
result['TFKFD'] = result['FD'] + result['FK']
result['FK:(FD+FK)'] = result['FK'] / (result['FD'] + result['FK'])
result
```

```python
# Most aggresive player (first kill/ (first death + first kill))
# Min. total average round played
average_round = stats['Rnd'].sum() / len(stats['Rnd'])

result = stats[stats['Rnd'] > average_round][['Player', 'Team','Agents', 'Rnd', 'FD', 'FK']]
result['TFKFD'] = result['FD'] + result['FK']
result['FK:(FD+FK)'] = result['FK'] / (result['FD'] + result['FK'])
result = result.sort_values('FK:(FD+FK)',ascending=False, ignore_index=True)
result
```
